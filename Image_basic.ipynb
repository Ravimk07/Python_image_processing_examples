# Create a black image

```python
# Create a black image
img = np.zeros((512,512,3), np.uint8)
plt.imshow(img)
plt.show()
```

# Draw a diagonal blue line with thickness of 5 px

```python
# Draw a diagonal blue line with thickness of 5 px
img = cv2.line(img,(0,0),(511,511),(0,0,255),5)
plt.imshow(img)
plt.show()
```

```python
# Draw a diagonal blue line with thickness of 5 px
img = cv2.line(img,(0,0),(size[1],size[0]),(0,0,255),5)
plt.imshow(img)
plt.show()
```

```python
# Drawing Rectangle
img = cv2.rectangle(img,(384,0),(510,128),(0,255,0),3)
plt.imshow(img)
plt.show()

```

```python
# Drawing Circle
img = cv2.circle(img,(247,63), 63, (0,0,255), -1)
plt.imshow(img)
plt.show()
```

# Adding Text to image

```python
# Adding Text to image
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img,'Ravi ',(10,300), font, 2,(255,255,255),2,cv2.LINE_AA)

plt.imshow(img)
plt.show()
```
